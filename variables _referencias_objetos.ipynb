```python

a = 3
b = a               # b es 3
print("b es ", b)

a = 3
b = a
a = 'spam'          # b es 3
print("b es ", b)
```

El valor A vale 3 y B es igual a A entonces b coge el valor de A, después a A se le asigna un nuevo valor que es 'spam' y B se queda con el valor inicial de A que es 3.

Los objetos `integer` no son mutables:


```python
a = 3
b = a
a = a + 2              # b es 3
print("b es ", b)
```

Es el mismo ejemplo que el anterior, la última asignación coge un nuevo resultado que en este caso es el 5, que sale de la expresión a + 2.  B quedaría con el valor inicial de A que es 3.

## OBJETOS MUTABLES

### Listas

```python

L = [2, 3, 4]	    # A mutable object
M = L			    # Make a reference to the same object
L[0] = 24 		    # An in-place change
print("L = ", L)    # [24, 3, 4]
print("M = ", M)    # [24, 3, 4]
```

M coge el valor de L, M y L tienen el mismo valor. L coge el primer valor de la serie (lista) y lo cambia por otro numero, en este caso 24. Entonces el valor 2 de L y M cambia a 24.

Copia de objetos en vez de referencia:


```python

L = [2, 3, 4]
M = L [:]		    # Make a copy of L (or list(L), copy.copy(L), etc.)
L[0] = 24

print("L = ", L)    # [24, 3, 4]

print("M = ", M)    # [2, 3, 4]  M is not changed
```

El cambio realizado a través de L no se refleja en M porque M hace referencia a una copia del valor al que L hace referencia, no al original

### Referencias compartidas e igualdad
- Operador `==` igualdad de valores.
- Operador `is` identidad de objetos.

```python

L = [1, 2, 3]
M = L 			    # M y L hacen referencia al mismo objeto

L == M 			    # True: Mismos valores => operador igualdad de valores
L is M 			    # True: operador identidad de objetos: compara las referencias (los punteros)
```

M y L tienen el mismo valor porque M coge el valor de L. is muestra que ambos objetos tengan el mismo valor por eso el resultado que devuelve es true.

```python
L = [1, 2, 3]
M = [1, 2, 3]	    # M y L hacen referencia a objetos diferentes

print(
    "L == M",
    L == M 			# True: Mismos valores
)

print(
    "L is M", 	    # False: operador identidad de objetos: compara las referencias (los punteros)
    L is M
)
```

L y M tienen el mismo valor por eso nos devuelve un true ==, pero is da false porque no apunta al mismo objeto

#### CACHE

Dado que los números enteros pequeños y las cadenas se almacenan en caché y se reutilizan, esto nos indica que hacen referencia al mismo objeto.


```python

X = 42
Y = 42		    # Deben ser dos objetos diferentes

X  == Y         # True
X is Y 		    # True: El mismo objeto, en cualquier caso: caché funcionando
```

X e Y tienen el mismo valor (42) pero tienen nombres distintos, esto quiere decir que apuntan a un mismo objeto.
Si usas objetos mutables son objetos distintos, si usas objetos inmutables son los mismos objetos.

Averiguar el numero de refencias a un objeto:


```python
import sys
sys.getrefcount(1)		# número de punteros a esta porción de memoria compartida
```

Este codigo cuenta cuantas variables están apuntando a un valor guardado en la memoria

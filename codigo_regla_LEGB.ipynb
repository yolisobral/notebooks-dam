## ÁMBITO GLOBAL

- Los nombres globales son variables asignadas en el nivel superior del archivo del módulo que las contiene.

- Los nombres globales solo deben declararse si se asignan dentro de una función.

- Se puede hacer referencia a los nombres globales dentro de una función sin necesidad de declararlos.

```python
x = 88 		 # Global x

def func():
	x = 99   # Local x: oculta la variable global, pero la queremos aquí.

func()
print(x)	 # Prints 88: no cambia
```

Se coge el valor global porque esta fuera de la función y existe en memoria, x = 99 no existe en memoria

```python
x = 88 			# Global x

def func():
	global x
	x = 99 		# Global x: definción exterior

func()
print("x =", x) # Prints 99
```

En este defines que x = 99 es global entonces cambia su valor 

## Regla LEGB 

```python
# Se puede hacer referencia a nombres globales dentro de una función sin necesidad de declararlos.

y, z = 1, 2         # Variables globales en el módulo

def all_global():
	global x        # Declarar variables globales asignadas
	x = y + z       # No es necesario declarar y, z: regla LEGB
                    # x existe ahora en el ámbito global con valor 3

all_global()
print('x = %d, y = %d, z = %d' % (x, y, z))
```

% se sustituye por los valores x, y, z

y, z tienen su valor fuera de la función, x es el único que coge un valor, en este caso 3 por la suma de y, z

## ÁMBITOS ANIDADOS


```python

x = 99                            # Nombre del ámbito global: no se utiliza

def f1():
    x = 88
    def f2():                     # Definición local que encierra
        print("x local = ", x)    # Referencia hecha en definición anidada
    f2()                          # f2 es una función temporal que solo existe durante la ejecución de la función f1 que la contiene (y solo es visible para el código dentro de ella).
f1()                              # Prints 88: definición local adjunta
print("x global = ", x)           # salida: x = 88; x sigue valiendo 99

# f2() NameError: el nombre 'f2' no está definido No se puede invocar a f2() desde el módulo principal
```

Por precedencia cuando se ejecuta f(2) coge el valor 88, f(1) coge el valor global

## Funciones de fábrica: Clausuras

Una **clausura** o una **función de fábrica**, la primera describe una **técnica de programación funcional** y la segunda denota un **patrón de diseño**.

El objeto función en cuestión almacena los valores en los ámbitos que la contienen, independientemente de si dichos ámbitos aún están presentes en la memoria.

```python
def f1():
	x = 88                 # ámbito de aplicación
	def f2():
		print("x =", x)    # Recuerda que x está dentro del ámbito de definición que lo contiene.
	return f2 	           # Devuelve f2 pero no lo llamas => f1() devuelve el objeto funcion con nombre (referencia) f2
action = f1()      		   # Mcrear, devolver función => acción ahora es una función: acción = f2
action() 				   # Llámalo ahora: prints 88  == f2()

# En Python, las funciones son objetos como cualquier otro elemento y pueden devolverse como valores de retorno desde otras funciones. 
# Lo más importante es que f2 recuerda la x del ámbito contenedor en f1, aunque f1 ya no esté activo.
```

Action va a buscar f(1) pero este ya deberia estar eliminado 

```python
def maker(n):
	def action(x):		# Acción de hacer y devolver
		return x ** n 	# La acción retiene n del ámbito de aplicación
	return action

f = maker(2)            # Pasar 2 al argumento n -> devolver x ** 2
f(3)                    # Pasar 3 a x, n recuerda "2" -> devolver 3 ** 2

f(4)		            # devolver 4 ** 2
```

El valor n vale 2 (guardada en memoria), f = action, f(3) se lo pasas a action

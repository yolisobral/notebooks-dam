Diccionarios
============

Los items se almacenan y recuperar por una clave, en vez de por su posicin en la estructura.

Es una tipo de agregacin en el cual el nombre de los items tiene ms significado que su posicin.

Los diccionarios pueden reemplazar muchos de los algoritmos de bsqueda y estructuras de datos que deberas implementar manualmente en lenguajes de bajo nivel.

Caractersticas:

* Se accede a los elementos por su clave o key, no por la posicin que ocupan.
* Son colecciones no ordenadas de objetos de cualquier tipo, y el tipo de los elementos que lo forman puede ser distinto.
* Poseen longitud variable y pueden anidarse.
* Son mapas mutables.
* Son tablas de referencias a objetos o hash tables.

```python
# D = {key: value}
D = {'spam': 2, 'ham': 1, 'eggs': 3}        # construir un diccionario
D
```

Tcnicamente, el orden es pseudoaleatorio; no es realmente aleatorio.

```python
D['spam']           # recuperar un valor su clave
```

```python
len(D)      # Numero de entradas del diccionario
# 3
```

```python
'ham' in D      # Key membership test alternative
# True
```

```python
list(D.keys())      # Crear una lista con las claves del diccionario
# ['eggs', 'spam', 'ham']
```

Es necesario encerrar el objeto que devuelve `keys()` en el constructor `list` en Python 3.X porque `keys()` en 3.X devuelve un objeto iterable en vez de una lista fsica.

## Cambiando diccionarios _in place_

```python
D['ham'] = ['grill', 'bake', 'fry']         # Modificar entrada (valor=lista)
D
# {'eggs': 3, 'spam': 2, 'ham': ['grill', 'bake', 'fry']}
```

```python
del D['eggs']               # Eliminar entrada
D
# {'spam': 2, 'ham': ['grill', 'bake', 'fry']}
```

Al contrario que las listas, cuando asignas una nueva clave en un diccionario (una que no haya sido asignada con anterioridad) creas una nueva entrada en el diccionario.

```python
D['brunch'] = 'Bacon'       # Agregar nueva entrada
D
# {'brunch': 'Bacon', 'spam': 2, 'ham': ['grill', 'bake', 'fry']}
```

### Mtodos de un objeto diccionario

#### `values`
El mtodo `values` devuelve todos los valores del diccionario.


```python
D = {'spam': 2, 'ham': 1, 'eggs': 3}
list(D.values())                
# [3, 2, 1]
```

#### `items` 
`items` devuelve todas las tuplas `(key,value)`.

```python
list(D.items())
# [('eggs', 3), ('spam', 2), ('ham', 1)]
```

#### `get`

Fetching una clave que no existe es normalmente un error, pero el mtodo `get` devuelve `None` como valor por defecto o un valor pasado como por defecto si la clave no existe.

```python
D['toast']      # KeyError: 'toast'   
```

```python
D.get('spam')
# 2
```

```python
print(D.get('toast'))       # Una clave que no existe
# None
```

```python
D.get('toast', 88)          # proporcionando un valor por defecto
# 88
```

#### `update`

El metodo `update` une las claves y valores de un diccionario en otro, sobreescribiendo valores de la misma clave si hay conflicto.

```python
# D = {'eggs': 3, 'spam': 2, 'ham': 1}
D2 = {'toast':4, 'muffin':5, 'ham': 10}
D.update(D2)
D
# {'eggs': 3, 'muffin': 5, 'toast': 4, 'spam': 2, 'ham': 10}
```

#### `pop`

`pop` elimina una clave de un diccionario y devuelve el valor asociado a la clave.

```python
D.pop('muffin')
# 5
D.pop('toast')
# 4
D
# {'eggs': 3, 'spam': 2, 'ham': 10}
```

## Ejemplos

Movie Database


```python
table = {'1975': 'Holy Grail', '1979': 'Life of Brian', '1983': 'The Meaning of Life'}
year = '1983'
movie = table[year]
movie
# 'The Meaning of Life'
```

```python
for year in table:                      # Los mismo que: for year in table.keys()
    print(year + '\t' + table[year])
```

Para cualquier diccionario D, `for key in D` funciona de la misma manera que `for key in D.keys()`

```python
for key in table:
    print(key + '\t' + table[key])
```

```python
for key in table.keys():
    print(key + '\t' + table[key])
```

```python
table = {'Holy Grail': '1975', 'Life of Brian': '1979', 'The Meaning of Life': '1983'}
table['Holy Grail']
```

```python
list(table.items())         # lista de tuplas (clave, valor)
```

#### Sintaxis de comprensin

```python
[title for (title, year) in table.items() if year == '1975']
```

```python
V = '1975'
[key for (key, value) in table.items() if value == V]
```

```python
[key for key in table.keys() if table[key] == V]
```

```python
list(K)
```

```python
V = D.values()              # Ditto para la vista valores e items
V
```

```python
list(V)
```

```python
D.items()
```

```python
K[0]            # TypeError: 'dict_keys' object is not subscriptable
```

```python
list(K)[0]
```

```python
D = dict(a=1, b=2, c=3)         # D = {'b': 2, 'c': 3, 'a': 1}
K = D.keys()
V = D.values()
del D['b']
D
```

```python
list(K)                 # Cambio reflejado en todas las vistas
# ['c', 'a']
```

```python
list(V)                 # Cambio reflejado en todas las vistas
# [3, 1]
```

## Vistas diccionario y sets

La vista objeto devuelta por el mtodo `keys` es como un `set` y soporta las operaciones con conjuntos, como la interseccin y la unin.

```python
K, V
```

```python
K | {'x': 4}            # La vista `keys` es como un `set` 
```

La vista `values` no es como un `set`.

```python
V & {'x': 4}            # TypeError: unsupported operand type(s) for &: 'dict_values' and 'dict'
```

La vista `items` genera pares (key, value) que son nicos y hashable (inmutables).

```python
V & {'x': 4}.values()   # TypeError: unsupported operand type(s) for &: 'dict_values' and 'dict_values'
```

## Ordenando claves

```python
Ks = D.keys()
Ks = list(Ks)
Ks.sort()
for k in Ks:
    print(k, D[k])
```

```python
D = {'b': 2, 'c': 3, 'a': 1}
Ks = D.keys()
for k in sorted(Ks): 
    print(k, D[k])
```

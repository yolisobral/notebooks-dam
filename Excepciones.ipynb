# Try/except/finally

"Unified try Example" del capítulo 34 Exception Coding Details (pag.1105) del libro Learning Python

### Primer caso: "EXCEPTION RAISED AND CAUGHT"

```python
sep = '' * 45 + '\n'
print(sep + 'EXCEPTION RAISED AND CAUGHT')
try:
    x= 'spam' [99]
except IndexError:
    print('except run')
finally:
    print('finally run')
print('after run')
```

'spam'[99] intenta acceder al índice 99 de la cadena 'spam', que no existe entonces lanza un IndexError.

El except IndexError captura la excepción y ejecuta su bloque.

Luego, el bloque finally SIEMPRE se ejecuta.

Finalmente, el programa continúa después del try y ejecuta print('after run').

### Segundo caso: "NO EXCEPTION RAISED"

```python
print(sep + 'NO EXCEPTION RAISED')
try:
    x='spam'[3]
except IndexError:
    print('except run')
finally:
    print('finally run')
print('after run')
```

'spam'[3] es 'm', no hay error.

Como no hay excepción, el bloque except no se ejecuta.

Aun así, el bloque finally sí se ejecuta siempre.

Luego, after run se imprime.

### Tercer caso: "NO EXCEPTION RAISED, WITH ELSE"

```python
print(sep + 'NO EXCEPTION RAISED, WITH ELSE')
try:
    x = 'spam'[3]
except IndexError:
    print('except run')
else:
    print('else run')
finally:
    print('finally run')
print('after run')
```

No hay error (igual que antes).

Como no hubo excepción, se ejecuta el bloque else.

Luego el finally (siempre se ejecuta).

Finalmente after run.

### Cuarto caso: “EXCEPTION RAISED BUT NOT CAUGHT”

```python
print(sep + 'EXCEPTION RAISED BUT NOT CAUGHT')
try:
    x=1 / 0
except IndexError:
    print('except run')
finally:
    print('finally run')
print('after run')
```

1 / 0 lanza una excepción ZeroDivisionError.

El bloque except solo atrapa IndexError, así que no la atrapa.

Antes de que el programa se detenga, el bloque finally se ejecuta (siempre se ejecuta).

Después, el programa se interrumpe y se muestra el rastreo del error (traceback).

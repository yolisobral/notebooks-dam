## Introducción a los Diccionarios en Python

Un diccionario es como una agenda telefónica o un glosario. En lugar de buscar una palabra por su posición (como la tercera palabra o el primer número), la buscas por el **nombre** que le corresponde.

## Conceptos Clave

**Almacén de Clave-Valor (Key-Value Store)**:

Un diccionario almacena información en pares. Cada par tiene una Clave (como el nombre de la persona en la agenda) y un Valor (como el número de teléfono de esa persona).

En Python, a esto se le llama **Mapping**.

**Clave** (`key`):

Es el identificador único que usas para buscar el valor. Piensa en ella como la etiqueta o el índice de un dato.

**Valor** (`value`):

Es el dato real que quieres almacenar o recuperar.

## Ventajas de los diccionarios

**Búsquedas muy rápidas**: Como accedes a los elementos directamente por su clave y no tienes que "recorrer" toda la lista, la búsqueda es extremadamente eficiente.

**No están ordenados**: A diferencia de una lista, los diccionarios no mantienen un orden estricto de los elementos

## Creación y Modificación de un Diccionario

Aquí tienes las formas más comunes de crear y manipular diccionarios:

## 1. Crear un Diccionario

### Sintaxis Literal

```python
d = {}                  # diccionario vacío
d = {'key': 'value'}    # diccionario con valores iniciales
```

Python 3.x Version ≥ 3.5

# También es posible desempaquetar uno o varios diccionarios con la sintaxis literal.
# Crea una copia superficial de otherdict.

d = {**otherdict}
# Además, actualiza la copia superficial con el contenido de yetanotherdict.

d = {**otherdict, **yetanotherdict}

### Comprensión de diccionario

```python
d = {k:v for k,v in [('key', 'value',)]}
```

### Clase integrada: `dict()`

```python
d = dict()                          # diccionario vacío
d = dict(key='value')               # argumentos de palabra clave explícitos
d = dict([('key', 'value')])        # pasando una lista de pares clave/valor
                                    # hacer una copia superficial de otro diccionario (¡solo posible si las claves son solo cadenas!)
d = dict(**otherdict)
```

## 2. Modificar un Diccionario

**Añadir o Actualizar** Elementos

Para añadir un par Clave-Valor, simplemente especificas la nueva Clave entre corchetes y le asignas un Valor.

- Si la clave no existe, se añade un nuevo par:

```python
d['newkey'] = 42
```

Si la clave ya existe, se actualiza (sobrescribe) el valor anterior:

```python
d['new_list'] = [1, 2, 3]
d['new_dict'] = {'nested_dict': 1}
```

Para eliminar un par Clave-Valor completo, usas la palabra clave del seguida `del` nombre del diccionario y la Clave que deseas eliminar:

```python
del d['newkey']
```
